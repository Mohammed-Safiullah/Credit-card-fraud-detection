# Feature Selection

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
#intiliaze kbest
kbest = SelectKBest(f_classif, k=20)
kbest.fit(X, y)

# Step 4: Get the indices of the top features
top_features_idx = kbest.get_support(indices=True)

# Step 5: Get the names of the top features
top_features_names = list(X.columns[top_features_idx])

# Step 6: Print the top features
print("The top feature(s) are:", top_features_names)
```

```python
feature_scores = list(zip(kbest.scores_,top_features_names))
sorted_feature_scores = sorted(feature_scores,reverse=True)


num_list = []
col_list = []
for i in range(len(sorted_feature_scores)):
    num_list.append((sorted_feature_scores[i])[0])
    col_list.append((sorted_feature_scores [i])[1])
```

```python
plt.bar(col_list,num_list)
plt.xticks(rotation=90)
plt.title('Feature Importance Using ANOVA')
```

```python
# use only selected features
X=X[top_features_names]
```

**Function Evualtion**

```python
# function for evaluation metrics precision, recall, f1 ROC-AUC score
from sklearn.metrics import roc_auc_score,roc_curve, accuracy_score, precision_score, recall_score, f1_score, classification_report,confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
def modelEvaluation(predictions, y_test_set, model_name):
    # Print model evaluation to predicted result    
    print("==========",model_name,"==========")
    print ("\nAccuracy on validation set: {:.4f}".format(accuracy_score(y_test_set, predictions)*100))    
    print ("\nOutput: \n", classification_report(y_test_set, predictions))
    print ("\nConfusion Matrix : \n", confusion_matrix(y_test_set, predictions))
    sns.heatmap(confusion_matrix(y_test_set, predictions),annot=True, fmt='g',cmap='viridis')
    plt.tight_layout()
    plt.show()
    results = [accuracy_score(y_test_set, predictions), precision_score(y_test_set, predictions, average='macro'),
              recall_score(y_test_set, predictions, average='macro'),f1_score(y_test_set, predictions, average='macro')]
    return results
```

###**One Hot Encoding**

```python
#conevreting the labels into one hot encoded labels
from keras.utils.np_utils import to_categorical   
y_one_hot = to_categorical(y, num_classes=2)
```

###**Data Spliting for Deep Learning Model**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1)
```

**Neural Network**

```python
#Neural Netwrok Architecture
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(20,)),
    keras.layers.Dense(32, activation=tf.nn.relu),
    keras.layers.Dense(32, activation=tf.nn.relu),
    keras.layers.Dense(2, activation=tf.nn.softmax),])
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1)
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

#**epochs**

```python
history = model.fit(X_train, y_train, epochs=10, batch_size=100, validation_data=(X_val, y_val))
```
